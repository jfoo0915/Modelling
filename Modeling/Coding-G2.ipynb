# 3 Methodology
## 3.1 Type of Simulation Used
- Dynamic simulation represents a system changes its behavior according to the time while static simulation represents a system at a particular time or the time has no role (Study.com, n.d.). Our self-checkout system is a dynamic simulation model. This is because the dynamic simulation model is a representation of a system that changes over time. In our system, the number of customers in the queue is the state variable. The number of customers in the queue waiting for checkout will change over time when there is a customer who arrives in the queue and when there is a customer who leaves the queue. Thus, it is a dynamic simulation model.<br><br>
 
- Beside that, the simulation model of our proposed self-checkout system is a probabilistic model. Deterministic simulation model can be used when the system does not contain any random variation (TKI.org, n.d.). Probabilistic is the process of expressing this uncertainty by giving the inputs as a probability distribution and specifying all random events that can affect the system (Goldsim, 2020). Hence, it includes the elements of randomness in our system. For instance, the arrival rate of customers is not constant throughout the day. Therefore, we need to take that into consideration.<br><br>

- Lastly, our proposed self-checkout system is also a discrete model. Discrete model is a method for modeling real-world systems that can be decomposed into a series of logically distinct processes that proceed autonomously over time (Barrett et al.,2008). Continuous simulation model is a type of simulation in which state variables change continuously over time (Leonelli, 2021). The output which is the average number of customers in the system is a discrete data that is produced because it allows us to measure the state variables such as the number of customers at the separate point of time. The number of customers will be changed only when the customers arrive or leave.<br><br>

## Case 1： Current Queue System (3 Manual Checkout Counters with 3 queues)

Firstly, a function called `make_system` that takes `lam` and `mu` as parameters and returns a `System` object with variables `lam`, `mu` and `duration` is written. `duration`, which is the number of time steps to simulate is set to 12 hours (operation hours of AEON from 10am - 10pm), expressed in minutes.

```python
def make_system(lam, mu):
    """Make a System object.
    
    lam: arrival rate, per minute
    mu: service completion rate, per minute
    
    returns: System object
    """
    # duration is 12 hours, expressed in minutes
    return System(lam = lam, mu = mu, duration =12*60)
```

This function is tested by creating a `System` object with `lam=1/3` and `mu=1/2.7`.

```python
interarrival_time = 3
service_time = 2.7

lam = 1 / interarrival_time
mu = 1 / service_time

system = make_system(lam, mu)
system
```

To simulate this scenario with two separate queues, 3 state variables (x1, x2, x3) are needed to keep track of customers in each queue.

An `update_func` that takes `x1`, `x2`, `x3`, `t`, and `system` as parameters and returns `x1`,`x2`, `x3` as return values is created.

`x1`: number of customers in queue 1
<br>`x2`: number of customers in queue 2
<br>`x3`: number of customers in queue 3
<br>`t`: time step (no. of minutes elapsed in the simulation)
<br>`system`: `System` object

If a customer is doing the checkout at the counter, `flip` should be used to decide whether they have complete te checkout. Also, `flip` should be used to decide if a new customer has arrived to the queue.

The `update_func` should return the number of customers for each queue at the end of the time step.

We assume that when a customer arrives, the customer will usually join the shorter queue.

```python
def update_func(x1, x2, x3, t, system):
    """Simulate 3 manual counters with 3 queues.
    
    x1: number of customers in queue 1
    x2: number of customers in queue 2
    x3: number of customers in queue 3
    t: time step
    system: System object
    """
    # if the first server is busy, check if it's done
    if x1 > 0 and flip(system.mu):
        x1 -= 1
            
    # if the second server is busy, check if it's done
    if x2 > 0 and flip(system.mu):
        x2 -= 1
            
    # if the third server is busy, check if it's done
    if x3 > 0 and flip(system.mu):
        x3 -= 1
    
    # check for an arrival
    if flip(system.lam):
        # join whatever queue that is shorter
        if x1 < x2 and x1 < x3:
            x1 += 1
        elif x2 < x1 and x2 < x3:
            x2 += 1
        else:
            x3 += 1
            
    return x1, x2, x3
```

To test if this function is working, this `update_func` is called by passing in the data:<br><br>
`x1=1`, `x2=1`, `x3=1`, `t=0` and the `System` object created. <br><br>
If the function is working properly, then different results will be shown in each execution.

```python
update_func(1, 1, 1, 0, system)
```

After that, we need to create a `run_simulation` function to works with `update_func` in order to run the simulation. The  `run_simulation` function will create a `TimeSeries` for the results obtained, which is the total number of customers in the queues.

```python
def run_simulation(system, update_func):
    """Simulate a 3 manual counters 3 queues system
    
    system: System object
    update_func: function object
    """
    x1, x2, x3 = 0, 0, 0
    results = TimeSeries()
    results[0] = x1 + x2 + x3
    
    for t in linrange(0, system.duration):
        x1, x2, x3 = update_func(x1, x2, x3, t, system)
        results[t+1] = x1 + x2 + x3

    return results
```

The simulation can be run with a single value of `lam`. The results are plotted.

```python
results = run_simulation(system, update_func)
plot(results)
decorate(xlabel='Time (min)', ylabel='Number of customers')
```

After the simulation, we can compute `L`, which is the average number of customers in the AEON checkout system, and `W`, which is the average time customers waiting in the queue for checkout.  `L` and `W` are related by Little's Law:

$L = \lambda W$

Where $\lambda$ is the arrival rate. The `compute_metrics` function with parameters `results` and `system` is used to compute them.

```python
def compute_metrics(results, system):
    """Compute average number of customers and waiting time.
    
    results: TimeSeries of queue lengths
    system: System object
    
    returns: L, W
    """
    L = results.mean()
    W = L / system.lam
    return L, W
```

`compute_metrics` fucntion is called with the `results` from the simulation.

```python
compute_metrics(results, system)
```

### Parameter sweep

Since the actual value of $\lambda$ is not known, a range of possibilities needs to sweep through, from 10% to 90% of the completion rate, $\mu$.  (If customers arrive faster than the completion rate, the queue grows without bound.  In that case the metrics `L` and `W` just depend on the operation hours of AEON United Point.)

An array of values for `lam` is created.

```python
num_vals = 101
lam_array = linspace(0.1*mu, 0.9*mu, num_vals)
lam_array
```

After that, the `sweep_lam` function is created, which will take an array of values for `lam`, a single value for `mu`, and the `update_func`.

For each value of `lam`, it should run a simulation, compute `L` and `W`, and store the value of `W` in a `SweepSeries`.

It should return the `SweepSeries`.

```python
def sweep_lam(lam_array, mu, update_func):
    """Run simulations with a range of values for `lam`
    
    lam_array: array of values for `lam`
    mu: probability of complete the checkout process for each time step
    update_func: passed along to run_simulation
    
    returns: SweepSeries of average wait time vs lam
    """
    sweep = SweepSeries()
    
    for lam in lam_array:
        system = make_system(lam, mu)
        results = run_simulation(system, update_func)
        L, W = compute_metrics(results, system)
        sweep[lam] = W
        
    return sweep
```

The `sweep_lam` function will generate a `SweepSeries`, which will sweep a range of values for `lam`, the results will be plotted and the average waiting time across all values of `lam` will also be printed.

### Analysis

The model chosen for this system is a common model in queueing theory, in part because many of its properties can be derived analytically.

In particular, the average time spent to checkout in the AEON United Point as a function of $\mu$ and $\lambda$ can be derived:

$W = 1 / (\mu - \lambda)$

The following function plots the theoretical value of $W$ as a function of $\lambda$.

```python
def plot_W(lam_array, mu):
    """Plot the theoretical mean waiting time.
    
    lam_array: array of values for `lam`
    mu: probability of completing the checkout process
    """
    W = 1 / (mu - lam_array)
    plot(lam_array, W, 'g-')
```

As a result of observation in other AEON branch that is implemented with self-checkout counters, it is found out that the service time for it is averagely 2 minutes.

```python
interarrival_time = 3
service_time = 2

lam = 1 / interarrival_time
mu = 1 / service_time

system2 = make_system(lam, mu)
system2
```

To simulate this scenario with two separate queues, 3 state variables (x1, x2, x3) are needed to keep track of customers in each queue.

An `update_func2` that takes `x1`, `x2`, `x3`, `t`, and `system2` as parameters and returns `x1`,`x2`, `x3` as return values is created.

`x1`: number of customers in queue 1
<br>`x2`: number of customers in queue 2
<br>`x3`: number of customers in queue 3
<br>`t`: time step (no. of minutes elapsed in the simulation)
<br>`system2`: `System` object

If a customer is doing the checkout at the counter, `flip` should be used to decide whether they have complete te checkout. Also, `flip` should be used to decide if a new customer has arrived to the queue.

The `update_func2` should return the number of customers for each queue at the end of the time step.

We assume that when a customer arrives, the customer will usually join the shorter queue.

```python
def update_func2(x1, x2, x3, t, system2):
    """Simulate 3 manual counters with 3 queues.
    
    x1: number of customers in queue 1
    x2: number of customers in queue 2
    x3: number of customers in queue 3
    t: time step
    system2: System object
    """
    # if the first server is busy, check if it's done
    if x1 > 0 and flip(system2.mu):
        x1 -= 1
            
    # if the second server is busy, check if it's done
    if x2 > 0 and flip(system2.mu):
        x2 -= 1
            
    # if the third server is busy, check if it's done
    if x3 > 0 and flip(system2.mu):
        x3 -= 1
    
    # check for an arrival
    if flip(system2.lam):
        # join whatever queue that is shorter
        if x1 < x2 and x1 < x3:
            x1 += 1
        elif x2 < x1 and x2 < x3:
            x2 += 1
        else:
            x3 += 1
            
    return x1, x2, x3
```

```python
def run_simulation2(system2, update_func2):
    """Simulate a 3 manual counters 3 queues system
    
    system: System object
    update_func: function object
    """
    x1, x2, x3 = 0, 0, 0
    results2 = TimeSeries()
    results2[0] = x1 + x2 + x3
    
    for t in linrange(0, system2.duration):
        x1, x2, x3 = update_func2(x1, x2, x3, t, system2)
        results2[t+1] = x1 + x2 + x3

    return results2
```

<b>Point estimates for Case 1 to Case 5</b>

```python
import math
```

```python
tValue = 2.132
n = 20
```

```python
mAWT1=c["AWT1"].mean()
sdAWT1=c["AWT1"].std()

mAWT2=c["AWT2"].mean()
sdAWT2=c["AWT2"].std()

mAWT3=c["AWT3"].mean()
sdAWT3=c["AWT3"].std()

mAWT4=c["AWT4"].mean()
sdAWT4=c["AWT4"].std()

mAWT5=c["AWT5"].mean()
sdAWT5=c["AWT5"].std()

# point estimates and confidence intervals
peAWT1 = [mAWT1-(tValue*(math.sqrt((sdAWT1**2)/n))),mAWT1+(tValue*(math.sqrt((sdAWT1**2)/n)))]
peAWT2 = [mAWT2-(tValue*(math.sqrt((sdAWT2**2)/n))),mAWT2+(tValue*(math.sqrt((sdAWT2**2)/n)))]
peAWT3 = [mAWT3-(tValue*(math.sqrt((sdAWT3**2)/n))),mAWT3+(tValue*(math.sqrt((sdAWT3**2)/n)))]
peAWT4 = [mAWT4-(tValue*(math.sqrt((sdAWT4**2)/n))),mAWT4+(tValue*(math.sqrt((sdAWT4**2)/n)))]
peAWT5 = [mAWT5-(tValue*(math.sqrt((sdAWT5**2)/n))),mAWT5+(tValue*(math.sqrt((sdAWT5**2)/n)))]
```

```python
dataPe={'Averange waiting time(s)':['lower interval','upper interval'],
        'Case 1':[peAWT1[0],peAWT1[1]],
        'Case 2':[peAWT2[0],peAWT2[1]],
        'Case 3':[peAWT3[0],peAWT3[1]],
        'Case 4':[peAWT4[0],peAWT4[1]],
        'Case 5':[peAWT5[0],peAWT5[1]]}

dfPe=pd.DataFrame(dataPe)
dfPe
```
